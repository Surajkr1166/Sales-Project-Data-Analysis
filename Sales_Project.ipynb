```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#upload dataset from the device
from google.colab import files
uploaded = files.upload()
```

```python
#read the csv file
sales=pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
sales
```

```python
#fetching null data from Amount col
sales[sales['Amount'].isnull()]
```

```python
sales.info()
```

```python
#Droping unwanted columns such as status, unnamed1
sales.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
sales
```

```python
#Delete all nan values from the data frame
sales.dropna(inplace=True)
sales
```

```python
#Rename the column Marital_Status to Marital
sales.rename(columns={'Marital_Status':'Marital'},inplace=True)
sales
```

```python
#change the values of Marital col 0-->single, 1--> married
sales['Marital'].replace({0:'Single',1:'Married'},inplace=True)
sales
```

```python
#change the values of Gender col F-->Female, M-->Male
sales['Gender'].replace({'F':'Female','M':'Male','f':'Female','m':'Male'},inplace=True)
sales
```

```python
#visualize the count of males and females based on gender column using seaborn using 2*2 subplot
fig=sns.set_theme(style="darkgrid", font_scale=1.1)

fig = plt.figure(facecolor='#E8F6F3', figsize=(12, 8))
fig.suptitle('Gender & Marital Status Analysis',fontsize=18,fontweight='bold',color='#0B5345')

#pie chart for the Gender Ratio
plt.subplot(2,2,1)
c=['#3498DB', '#E74C3C']
ex=[0, 0.08]
plt.pie(sales['Gender'].value_counts(),labels=sales['Gender'].value_counts().index,autopct='%1.1f%%',colors=c,shadow=True,startangle=90,explode=ex,textprops={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'},wedgeprops={'edgecolor': 'white', 'linewidth': 2},radius=1,pctdistance=0.6)
plt.title('Gender Ratio', fontsize=14, fontweight='bold', color='#1A5276')

#countplot for the Gender count
plt.subplot(2, 2, 2)
ax = sns.countplot(x='Gender',data=sales,palette='coolwarm',width=0.5)
ax.set_facecolor('#FEF9E7')
plt.title('Gender Count', fontsize=14, fontweight='bold', color='#7D6608')
for container in ax.containers:
    ax.bar_label(container, fontsize=10, fontweight='bold', color='black')

#countplot for the martial status count
plt.subplot(2, 2, 3)
ax = sns.countplot(x='Marital',data=sales,palette='Set2',width=0.5)
ax.set_facecolor('#FDEDEC')
plt.title('Marital Status Count', fontsize=14, fontweight='bold', color='#922B21')
for container in ax.containers:
    ax.bar_label(container, fontsize=10, fontweight='bold', color='black')

#pie plot for the martial status ratio
plt.subplot(2, 2, 4)
marital_counts = sales['Marital'].value_counts()

plt.pie(marital_counts,labels=marital_counts.index,autopct='%1.1f%%',colors=['#1ABC9C', '#F4D03F'],explode=[0, 0.08],startangle=90,shadow=True,textprops={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'},wedgeprops={'edgecolor': 'white', 'linewidth': 2})
plt.title('Marital Status Ratio', fontsize=14, fontweight='bold', color='#0E6251')

fig.savefig('fig.jpeg')
plt.tight_layout()
plt.show()


```

****Conclusion: Gender & Marital Status Analysis****

From the above visualizations, we can conclude that the dataset is female dominant. Around 69.7% of the customers are female, while 30.3% are male. This indicates that female customers contribute more to the overall customer base.

The gender count bar chart also supports this observation, where the number of female customers is significantly higher than male customers. Hence, marketing strategies can be more focused on female customers to maximize engagement and sales.

In terms of marital status, single customers form the majority with approximately 58%, whereas married customers account for about 42% of the total population. This shows that single individuals are more active or more represented in the dataset.

Overall, the analysis suggests that female and single customers are the dominant groups, and businesses can design targeted campaigns, offers, and products keeping these customer segments in mind.

****Conclusion: Sales Amount Analysis****

From the above visualizations, it is clear that female customers contribute significantly more to total sales amount compared to male customers. This indicates that females are the primary revenue-generating segment in the dataset.

The age group 26–35 years contributes the highest total sales, followed by the 36–45 years age group. This shows that customers in the working-age category are more active buyers and have higher purchasing power.

State-wise analysis reveals that Uttar Pradesh generates the highest sales amount, followed by Maharashtra, Karnataka, and Delhi. These states represent major markets and should be prioritized for business expansion and targeted marketing strategies.

From the occupation analysis, customers working in the IT sector, Aviation, Banking, and Healthcare contribute the most to sales. This suggests that professionals with stable income have higher spending capacity.

Product category analysis shows that Food, Clothing & Apparel, Footwear & Shoes, and Electronics & Gadgets are the top revenue-generating categories. These categories have high demand and play a crucial role in overall sales performance.

Finally, the marital status analysis indicates that single customers generate more total sales amount than married customers, showing higher spending behavior among single individuals.

Overall, the analysis highlights that female, single, working-age customers from major states and professional occupations are the key contributors to sales, and businesses can use these insights to design better marketing and sales strategies
